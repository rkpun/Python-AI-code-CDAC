```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('datasets/mtcars.csv')
df
```

```python
df.shape
```

```python
# Seperate input and output variable
# disp, hp, wt
x=df[['disp','hp','wt']]
#mpg
y=df['mpg']
```

```python
x
```

```python
y
```

```python
x.corrwith(y)
```

```python
# Build the Model
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(x,y)
```

```python
regressor.score(x,y)
```

```python
regressor.coef_ ,regressor.intercept_
```

```python
#Predict
disp=221
hp=102
wt=2.91
new=pd.DataFrame({
    'disp':[221],
    'hp':[102],
    'wt':[2.91]
})
```

```python
new
```

```python
y_pred=regressor.predict(new)
```

```python
y_pred
```

```python
# Visualize
y_pred=regressor.predict(x)
y_pred
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
plt.scatter(x['disp'],y,color= 'green')
plt.subplot(1,3,2)
plt.scatter(x['hp'],y,color= 'orange')
plt.subplot(1,3,3)
plt.scatter(x['wt'],y,color= 'blue')
```

#### Adaboost the regression 

```python
from sklearn.ensemble import AdaBoostRegressor
reg_adb= AdaBoostRegressor(regressor, n_estimators=13, random_state=0)
```

```python
reg_adb.fit(x,y)
```

```python
y_p=reg_adb.predict(x)
```
