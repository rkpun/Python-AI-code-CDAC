```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('datasets/petrol_consumption.csv')
df
```

```python
x=df.drop('Petrol_Consumption',axis=1)   
y= df['Petrol_Consumption']   # Output Label
```

```python
# Cross Validation
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,random_state=0)
```

```python
# Build The model
from sklearn.ensemble import RandomForestRegressor
reg_rf= RandomForestRegressor(random_state=0,n_estimators=20)
```

```python
reg_rf.fit(x_train, y_train)
```

```python
y_pred=reg_rf.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
#Checking the Accuracy
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
reg_rf.estimators_
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(16,12))
plot_tree(reg_rf.estimators_[19], fontsize=6, feature_names=x.columns,filled=True, precision=2, rounded=True);
```

```python
result=pd.DataFrame({
    'Actual': y_test,
    'Predicted':y_pred
})
result
```
