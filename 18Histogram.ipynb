```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
num =[np.random.randint(0,100) for x in range(1000)]
```

```python
print(num)
```

```python
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Count')
plt.xticks(range(0,101,10))
plt.grid()
plt.hist(num,bins=range(0,101,10))
```
