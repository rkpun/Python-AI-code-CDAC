```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('images/lenna.png', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
img
```

```python
#Sequential Colormap
plt.figure(figsize=(16,9))
plt.subplot(2,4,1)
plt.title('Gray')
plt.imshow(img,cmap='gray')
plt.subplot(2,4,2)
plt.title('Blues')
plt.imshow(img,cmap='Blues')
plt.subplot(2,4,3)
plt.title('Greens')
plt.imshow(img,cmap='Greens')
plt.subplot(2,4,4)
plt.title('Reds')
plt.imshow(img,cmap='Reds')
plt.subplot(2,4,5)
plt.title('Oranges')
plt.imshow(img,cmap='Oranges')
plt.subplot(2,4,6)
plt.title('PuRd')
plt.imshow(img,cmap='PuRd')
```

```python
#Diverging Colormaps
plt.figure(figsize=(16,9))
plt.subplot(2,4,1)
plt.title('Seismic')
plt.imshow(img,cmap='seismic')
plt.subplot(2,4,2)
plt.title('coolwarm')
plt.imshow(img,cmap='coolwarm')
plt.subplot(2,4,3)
plt.title('bwr')
plt.imshow(img,cmap='bwr')
plt.subplot(2,4,4)
plt.title('spectral')
plt.imshow(img,cmap='Spectral')
plt.subplot(2,4,5)
plt.title('PiYG')
plt.imshow(img,cmap='PiYG')
plt.subplot(2,4,6)
plt.title('BrBG')
plt.imshow(img,cmap='BrBG')
```

```python
#Cyclic Colormaps
plt.subplot(1,4,1)
plt.title('hsv')
plt.imshow(img,cmap='hsv')
plt.subplot(1,4,2)
plt.title('turbo')
plt.imshow(img,cmap='turbo')
plt.subplot(1,4,3)
plt.title('twilight')
plt.imshow(img,cmap='twilight')
plt.subplot(1,4,4)
plt.title('twilight_shifted')
plt.imshow(img,cmap='twilight_shifted')
```

```python
#Qualatative Colormaps
plt.subplot(1,4,1)
plt.title('tab10')
plt.imshow(img,cmap='tab10')
plt.subplot(1,4,2)
plt.title('Set2')
plt.imshow(img,cmap='Set2')
plt.subplot(1,4,3)
plt.title('viridis')
plt.imshow(img,cmap='viridis')
plt.subplot(1,4,4)
plt.title('inferno')
plt.imshow(img,cmap='inferno')
```
