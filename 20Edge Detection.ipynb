```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('images/lenna.png',0)
```

```python
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
combined_sobel=cv2.add(sobelx,sobely)
```

```python
final_sobel= np.hstack([sobelx,sobely,combined_sobel])
cv2.imshow('Combined Image', final_sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(142)
plt.title('Sobel X')
plt.imshow(abs(sobelx), cmap='gray')
plt.subplot(143)
plt.title('Sobel Y')
plt.imshow(abs(sobely), cmap='gray')
plt.subplot(144)
plt.title('Final Image')
plt.imshow(abs(combined_sobel), cmap='gray')
```

```python
# Canny Edge detection 
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('images/houseright.jpg',0)
edges=cv2.Canny(img,50,150)
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.title('Canny')
plt.imshow(edges, cmap='gray')
```
