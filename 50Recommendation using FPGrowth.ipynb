```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
```

```python
# Create the data set
dataset=[['Apple','Beer','Rice','Chicken'],
         ['Apple','Beer','Rice'],
         ['Apple','Beer'],
         ['Apple','Pear'],
         ['Milk','Beer','Rice','Chicken'],
         ['Milk','Beer','Rice'],
         ['Milk','Beer'],
         ['Milk','Pear']
        ]
```

```python
dataset
```

```python
te =TransactionEncoder()
encoded= te.fit_transform(dataset)
```

```python
encoded
```

```python
te.columns_
```

```python
df=pd.DataFrame(encoded, columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
# find frequent itemset
freq_item = fpgrowth(df,min_support=0.25, use_colnames=True)
```

```python
freq_item
```

```python
# Form the rules
rules= association_rules(freq_item, metric='confidence', min_threshold=0.5)
```

```python
rules
```

```python
rules=rules[['antecedents','consequents','support','confidence']]
```

```python
rules
```

```python
# Sort the Rules
rules.sort_values(by=['confidence','support'], ascending=False)
```

```python
# Filter the rules
nrules=rules[rules['confidence']>0.5]
```

```python
nrules.shape
```

```python
nrules
```

```python
nrules[nrules['antecedents'] == {'Rice'}] ['consequents']
```

```python
nrules[nrules['antecedents'] == {'Apple', 'Beer'}] ['consequents']
```

```python
#Export the rules
nrules.to_csv('rules.csv', index=False)
```
