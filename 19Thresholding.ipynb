```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('images/sachin.jpg',0)
```

```python
new_img= img >127
inv_img= img <127
```

```python
new_img
```

```python
plt.figure(figsize=(16,9))
plt.subplot(131)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(132)
plt.title('Binary Image')
plt.imshow(new_img, cmap='gray')
plt.subplot(133)
plt.title('Inverse Binary Image')
plt.imshow(inv_img, cmap='gray')
```
