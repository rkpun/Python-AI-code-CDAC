```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img1= cv2.imread('images/input1.png')
img2= cv2.imread('images/input2.png')
dest_and = cv2.bitwise_and(img2,img1, mask=None)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.title('Image1')
plt.imshow(img1)
plt.subplot(1,3,2)
plt.title('Image2')
plt.imshow(img2)
plt.subplot(1,3,3)
plt.title('Bitwise AND Image')
plt.imshow(dest_and)
```
